2. What is the time complexity of the following code:
```
int a = 0;
for (i = 0; i < N; i++)
{
    for (j = N; j > i; j--)
    {
        a = a + i + j;
    }
}
```

```
☑:  approx  O(N*N)
```

3. What is the time complexity of the following code:
```
int i, j, k = 0;
for (i = n / 2; i <= n; i++)
{
    for (j = 2; j <= n; j = j * 2)
    {
        k = k + n / 2;
    }
}
```
