1.Write a C++ program to convert Primitive type to Complex type.
Example -
 ```

{
    Complex c1;
    Int x=5;
    c1=x;
    return 0;
}
```

```
#include <iostream>
using namespace std;

class Complex
    {
        private:
            float real;
            float imag;

        public:
            Complex()
                {
                    real = 0;
                    imag = 0;
                }

            Complex(int x)
                {
                    real = x;
                    imag = 0;
                }

            void display() const
                {
                    cout << real << " + " << imag << "i" << endl;
                }
};

int main()
    {
        Complex c1;
        int x = 5;
        c1 = x;
        cout << "Complex number: ";
        c1.display();
        return 0;
    }
```
