1. Write a C++ program to demonstrate the use of try, catch block with the argument as an integer and string using multiple catch blocks.

```
#include<iostream>
using namespace std;
#include <string>

class A
    {
        private:
            int a;
            string* b;
        public:

            A()
                {
                    a = -1;
                    b = nullptr;
                }

            void _set(int a, string b)
                {
                    try
                        {
                            if(a<0)
                                {
                                    throw("Entered number less than 0, called default value: " ,a);
                                }
                            if(b.empty())
                                {
                                    throw "String cannot be empty";
                                }

                            this->a = a;
                            this->b = new string(b);
                        }

                    catch(const int e)
                        {
                            cout << "Exception caught:" << e << endl;
                        }

                    catch(const char* errorMessage)
                        {
                            cout << "Exception caught: " << errorMessage << endl;
                        }

                }
             void display()
                {
                    cout << "Result: Integer = " << a << ", String = " << *b << endl;
                }

             ~A()
                {
                    delete b;
                }
    };

int main()
    {
        cout<< "Required two arguments :- "<<endl;
        int x;
        string y;
        cout<< "Enter First argument as number: ";
        cin >> x;
        cout<< endl << "Enter Second argument as string: ";
        cin >> y;
        cin.ignore();
        getline(cin, y);
        cout<< endl ;

        A obj;
        obj._set(x, y);
        obj.display();
        return 0;
    }


```

8. Write a C++ program to accept a password and throw an exception if the password has less than 6 characters or does not contain a digit or does not contain any special character or does not contain any capital letter.

```
#include<iostream>
using namespace std;
#include<cstring>
class Password
{
    private:
        char* passcode;
        int size;

    public:
        Password()
            {
                passcode = nullptr;
                size = 30;
            }
        ~Password()
            {
                if(passcode)
                    {
                        free(passcode);
                    }
            }

           void _set()
                {
                    try
                        {
                            passcode = (char*)malloc(size* sizeof(char));
                            if (passcode == nullptr)
                                {
                                    cout << "Memory allocation failed!" << endl;
                                    return;
                                }
                             cout<< "Enter passcode: ";

                            cin.getline(passcode, size);

                            if (strlen(passcode) > 6)
                                {
                                    for(int i= 0; passcode[i]!='\0'; i++)
                                        {
                                            if (isdigit(passcode[i]) || ispunct(passcode[i]) || isupper(passcode[i]))
                                                {
                                                    throw("Invalid password ! digits and symbol are not allowed ");
                                                }
                                            else
                                                {
                                                    cout<<passcode[i];
                                                }
                                        }
                                     cout<< endl<<"password accepted";
                                 }
                          else
                            {
                                throw("Invalid Password ! It must have length more than 6 ");
                            }
                    }
                catch(const char* error_msg)
                    {
                        cout << "Exception caught: " << error_msg << endl;
                    }
             }
};

int main()
{
    Password u;
    u._set();
    return 0;
}



```
