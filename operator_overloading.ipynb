1. Define a class Complex with appropriate instance variables and member functions.
Overload following operators
a. << insertion operator
b. >> extraction operator

```
#include <iostream>
using namespace std;

class Complex
{
private:
    int a;

public:
    Complex()
    {
        a = 1;
    }

    Complex(int x)
    {
        a = x;
    }


    friend istream &operator>>(istream &in, Complex &obj);
    friend ostream &operator<<(ostream &show, const Complex &obj);
};

istream &operator>>(istream &in, Complex &obj)
    {
        cout << "Enter a number: ";
        in >> obj.a;
        return in;
    }

ostream &operator<<(ostream &show, const Complex &obj)
    {
        show << "The current value of a is: " << obj.a;
        return show;
    }

int main()
    {
        Complex c(41);
        cout << c << endl;

        cin >> c;
        cout << c << endl;

        return 0;
    }
```

3. Overload subscript operator [] that will be useful when we want to check for an index out of bound.

```

#include<iostream>
using namespace std;

class subscript
{
    private:
        int arr[100];
        const int size =100;

    public:
        subscript(int x = 1)
            {
                arr[10] = x;
            }
         void setData(int position, int value)
            {
                if(position > size )
                    {
                       cout << "Index is out of bound";
                       exit(0);
                    }
                 else
                    {
                        arr[position] = value;
                    }
            }
         void display()
            {
               cout<<"The value of arr[10] = "<<arr[10]<<endl;
            }

         int operator[](int position)
            {
                if(position >= size)
                    {
                        cout << "Index is out of bound" << endl;
                        exit(0);
                    }
                cout << "[] operator called" << endl;
                return arr[position];
         }
};

int main()
{
    subscript s;
    s.setData(10,3);
    s.display();
    cout << "Accessing arr[10]: " << s[10] << endl;
    return 0;
}



```

7. Create an Integer class and overload logical not operator for that class.

```
#include<iostream>
using namespace std;

class Integer
    {
        private:
            int x;

        public:
            Integer(int x)
                {
                    this->x = x;
                }
            Integer operator!()
                {
                    cout << "Called Logical NOT!" << endl;
                    return Integer(!(this->x));
                }

            void display()
                {
                    cout << "Value: " << x << endl;
                }
    };

int main()
    {
        Integer i(10);
        i.display();
        Integer result = !i;
        result.display();
        return 0;
    }

```

```


#include<iostream>
using namespace std;

class Coordinate
{
    private:
        int x, y, z;

    public:
        Coordinate(int x = 0, int y = 0, int z = 0)
            {
                this->x = x;
                this->y = y;
                this->z = z;
            }

        void display() const
            {
                cout << "Coordinate: (" << x << ", " << y << ", " << z << ")" << endl;
            }

         Coordinate& operator,(Coordinate& z)
            {
                return z;
            }
};

int main()
    {
        Coordinate c1(2, 4, 1), c2(1, 2, 3), c3;

        c3 = (c1, c2);

        cout << "c1: ";
        c1.display();

        cout << "c2: ";
        c2.display();

        cout << "c3 (after c3 = (c1, c2)): ";
        c3.display();

        return 0;
    }


```

10. Create a Distance class having 2 instance variable feet and inches. Also create default constructor and parameterized constructor takes 2 variables . Now overload () function call operator that takes 3 arguments a , b and c and set feet = a + c + 5 and inches = a+b + 15.

```

#include<iostream>
using namespace std;

class Distance
    {
        private:
            int feet, inches;

        public:
            Distance()
                {
                    feet = 0;
                    inches = 0;
                }
            Distance(int x, int y)
                {
                   feet = x;
                   inches = y;
                }

           Distance operator()(int a, int b, int c)
                {
                    feet = a + c + 5;
                    inches = a+b + 15;

                    return *this;
                }

           void display()
                {
                    cout <<" feet"<< feet<< endl << "inches"<< inches;
                }
    };

int main()
    {
        Distance d1(1, 2), d2, d3;
        cout << "Initial Values:" << endl;
        d1.display();
        cout << endl;
        d2.display();
        cout << endl;
        d3 = d2(3, 4, 2);
        cout << "After Overloaded Operator Call:" << endl;
        d3.display();
        return 0;
    }




```

function that will print marks. We know that we can access member functions using (.) dot operator. Now you need to overload (->) arrow operator to access that function.

```
#include<iostream>
using namespace std;

class Marks
{
    private:
        int marks;

    public:
        Marks(int x)
            {
                marks = x;
            }

        void display()
            {
                cout<< "Marks Entered is: "<< marks<<endl;
            }

        Marks* operator->()
            {
                return this;
            }
};

int main()
{
    Marks m(98);
    m.display();
    m->display();
}


```
