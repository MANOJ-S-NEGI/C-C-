6. Write a C++ program to demonstrate how a common friend function can
be used to exchange the private values of two classes. (Use call by
reference method).

```

#include <iostream>
using namespace std;

class ClassB;

class ClassA
{
    private:
        int valueA;

    public:
        ClassA(int val) : valueA(val) {}

        void display() const
            {
                cout << "ClassA value: " << valueA << endl;
            }

        friend void exchangeValues(ClassA& objA, ClassB& objB);
};

class ClassB
{
    private:
        int valueB;

    public:
        ClassB(int val) : valueB(val) {}

    void display() const
        {
            cout << "ClassB value: " << valueB << endl;
        }

        friend void exchangeValues(ClassA& objA, ClassB& objB);
};

void exchangeValues(ClassA& objA, ClassB& objB)
    {
        int temp = objA.valueA;
        objA.valueA = objB.valueB;
        objB.valueB = temp;
    }

int main()
{
    ClassA objA(10);
    ClassB objB(20);

    cout << "Before exchanging values:" << endl;
    objA.display();
    objB.display();

    exchangeValues(objA, objB);

    cout << "\nAfter exchanging values:" << endl;
    objA.display();
    objB.display();

    return 0;
}


```

1. Create a class FLOAT that contains one float data member. Overload all the four arithmetic operators so that they can operate on the objects of FLOAT.

```
// Create a class FLOAT that contains one float data member. Overload all the four arithmetic operators so that they can operate on the objects of FLOAT.
#include <iostream>
using namespace std;

class FLOAT {
private:
    float value;

public:
    FLOAT(float v = 0.0) : value(v) {}

    FLOAT operator+(const FLOAT other)
        {
            return FLOAT(value + other.value);
        }

    FLOAT operator-(const FLOAT other)
        {
            return FLOAT(value - other.value);
        }

    FLOAT operator*( FLOAT other)
        {
            return FLOAT(value * other.value);
        }

    FLOAT operator/(const FLOAT other)
        {
            if (other.value != 0)
                {
                    return FLOAT(value / other.value);
                }
            else
                {
                    cout << "Error: Division by zero!" << endl;
                    return FLOAT(0);
                }
        }

    void display() const
        {
            cout << value << endl;
        }
};

int main()
{
    FLOAT num1(10.5), num2(2.5);

    FLOAT result_add = num1 + num2;
    cout << "Addition: ";
    result_add.display();

    FLOAT result_sub = num1 - num2;
    cout << "Subtraction: ";
    result_sub.display();

    FLOAT result_mul = num1 * num2;
    cout << "Multiplication: ";
    result_mul.display();

    FLOAT result_div = num1 / num2;
    cout << "Division: ";
    result_div.display();

    return 0;
}

```

2. Define a class Rectangle and overload area function for different types of data type.

```
#include <iostream>
using namespace std;

class Rectangle {
private:
    float length, width;

public:
    Rectangle() : Rectangle(0, 0) {}

    Rectangle(float l, float w) : length(l), width(w) {}

    float area()
        {
            return length * width;
        }

    int area(int l, int w)
        {
            return l * w;
        }

    double area(double l, double w)
        {
            return l * w;
        }
};

int main() {
    Rectangle rect(5.5, 4.5);

    cout << "Area (float): " << rect.area() << endl;
    cout << "Area (int): " << rect.area(5, 4) << endl;
    cout << "Area (double): " << rect.area(5.5, 4.5) << endl;
    return 0;
}


```

8. Write a C++ program to find the factorial of a number using copy constructor

```
//Write a C++ program to find the factorial of a number using copy constructor
#include<iostream>
using namespace std;

class Base
{
    private:
        int fact = 1;
        int a;
        int copy_fact =1;

    public:
        Base(int x):a(x)
            {
                for(int i = 1; i<=x; i++)
                    {
                        fact = fact*i;
                    }
            }
        void display()
            {
                cout<< "factorial : "<<fact<<endl;
                cout<<"copied  factorial : "<< copy_fact<<endl;
            }

         Base(const Base& b)
            {
                this->copy_fact = b.fact;
            }
};

int main()
{
    Base obj1(5);
    obj1.display();

    Base obj2 = obj1;
    obj2.display();
    obj1.display();
    return 0;
}

```

9. Write a C++ program to calculate the area of triangle, rectangle and circle using constructor overloading. The program should be menu driven.

```
#include<iostream>
#include<cmath>
using namespace std;

class Area
{
    private:
        float area;

    public:
        Area(): area(0) {}
        Area(float l, float b)
        {
            area = l * b;
        }

        Area(float r)
        {
            area = 3.14 * pow(r, 2);
        }
        Area(float b, float h, float e)
        {
            area = (e * b * h) / 2;
        }
        void display()
        {
            cout << area << endl;
        }
};

int main()
{
    int choice;
    float length, breadth, radius, base, height;

    cout << "Menu: \n 1. Rectangle \n 2. Circle \n 3. Triangle \n";
    cout << "Enter your choice: ";
    cin >> choice;

    switch(choice)
    {
        case 1:
            cout << "Enter length: ";
            cin >> length;
            cout << "Enter breadth: ";
            cin >> breadth;
            cout << "Area of Rectangle: ";
            {
                Area a(length, breadth);
                a.display();
            }
            break;

        case 2:
            cout << "Enter radius: ";
            cin >> radius;
            cout << "Area of Circle: ";
            {
                Area a(radius);
                a.display();
            }
            break;

        case 3:
            cout << "Enter base: ";
            cin >> base;
            cout << "Enter height: ";
            cin >> height;
            cout << "Area of Triangle: ";
            {
                Area a(base, height, 1.0);
                a.display();
            }
            break;

        default:
            cout << "Invalid choice!" << endl;
    }

    return 0;
}

```

10. Create a C++ class for player objects with the following attributes: player no., name, number of matches and number of goals done in each match. The number of matches varies for each player. Write a parameterized constructor which initializes player no., name, number of subjects and creates an array for number of goals and number of matches dynamically.


```

#include <iostream>
#include <string>
using namespace std;

class Player
{
    private:
        int player_no;
        string player_name;
        int number_of_matches;
        int* goals_per_match;

    public:
        Player(int num, string name, int matches, int* goals)
            {
                player_no = num;
                player_name = name;
                number_of_matches = matches;

                goals_per_match = new int[matches];
                for (int i = 0; i < matches; i++)
                    {
                        goals_per_match[i] = goals[i];
                    }
            }
        ~Player()
            {
                delete[] goals_per_match;
            }

        void display()
            {
                cout << "Player No: " << player_no << endl;
                cout << "Player Name: " << player_name << endl;
                cout << "Number of Matches: " << number_of_matches << endl;
                cout << "Goals per Match: ";

                for (int i = 0; i < number_of_matches; i++)
                    {
                        cout << goals_per_match[i] << " ";
                    }
                cout << endl;
            }
};

int main()
{
    int entry;
    cout << "Number of entries you want to store: ";
    cin >> entry;

    Player* players[entry];

    for (int i = 0; i < entry; i++)
     {
        int player_no, matches;
        string player_name;

        cout << "Enter player number: ";
        cin >> player_no;
        cin.ignore();
        cout << "Enter player name: ";
        getlines(cin,player_name);
        cout << "Enter number of matches: ";
        cin >> matches;

        int* goals = new int[matches];
        cout << "Enter goals for each match: ";
        for (int j = 0; j < matches; j++)
            {
                cin >> goals[j];
            }

        players[i] = new Player(player_no, player_name, matches, goals);

        delete[] goals;
    }

    cout << "\nPlayer Information:\n";
    for (int i = 0; i < entry; i++)
        {
            players[i]->display();
            delete players[i];
        }

    return 0;
}


```
