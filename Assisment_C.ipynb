7. Write a program to find the position of first 1 in LSB.

```
#include <stdio.h>
int LSB(int n)
{
  printf(" n is : %d\n", n);
  for (int position = 0; n > 0; position++)
    {
      if (n & 1)
        {
          return position;
        }
      n >>= 1;
      printf("now n is : %d\n", n);
    }
  return -1;
}

int main() {
    int number;
    printf("Enter a number: ");
    scanf("%d", &number);
    int position = LSB(number);

    if (position >= 0) {
        printf("The position of the first '1' in the LSB of %d is: %d\n", number, position);
    } else {
        printf("No '1' found in the binary representation of %d.\n", number);
    }
    return 0;
}

```

```
#include <stdio.h>

int Logic(int n, int m)
{
    int lcm = 1;
    
    for (int i = 2; i <= 10; i++)
        {
            int factorFound = 0;

            if (m % i == 0)
                {
                    m = m / i;           
                    factorFound = 1;
                }

            if (n % i == 0)
                {
                    n = n / i;           
                    factorFound = 1;
                }

            if (factorFound)
                {
                    lcm = lcm * i;
                    i--;
                }
        }
    lcm = lcm * m * n;
    return lcm;
}

int main() {
    int alpha, beta;
    printf("Enter 1st number:\n");
    scanf("%d", &alpha);

    printf("Enter 2nd number:\n");
    scanf("%d", &beta);

    int lcm = Logic(alpha, beta);
    
    printf("LCM of %d and %d is %d\n", alpha, beta, lcm);
    return 0;
}


```

* For the next 100 units Rs. 1.20/unit
* For units above 250 Rs. 1.50/unit
* An additional surcharge of 20% is added to the bill.

```
#include <stdio.h>

int main()
{
    int units;
    float bill = 0.0;

    printf("Enter the electricity units consumed: ");
    scanf("%d", &units);

    switch (units > 250)
        {
            case 1:
                bill += (units - 250) * 1.50;
                units = 250;
        }

    switch (units > 150)
        {
            case 1:
                bill += (units - 150) * 1.20;
                units = 150;
        }

    switch (units > 50)
        {
            case 1:
                bill += (units - 50) * 0.75;
                units = 50;
        }

    bill += units * 0.50;

    bill += bill * 0.20;

    printf("Total electricity bill: Rs. %.2f\n", bill);

    return 0;
}

```

```
# include <stdio.h>

int Logic(int n, int m)
{
    int lcm = 1;

    for (int i = 2; i <= 10; i++)
        {
            int factorFound = 0;

            if (m % i == 0)
                {
                    m = m / i;           
                    factorFound = 1;
                }

            if (n % i == 0)
                {
                    n = n / i;           
                    factorFound = 1;
                }

            if (factorFound)
                {
                    lcm = lcm * i;
                    i--;
                }
        }
    lcm = lcm * m * n;
    return lcm;
}

int main() {
    int alpha, beta;
    printf("Enter 1st number:\n");
    scanf("%d", &alpha);

    printf("Enter 2nd number:\n");
    scanf("%d", &beta);

    int lcm = Logic(alpha, beta);

    printf("LCM of %d and %d is %d\n", alpha, beta, lcm);
    return 0;
}
```

8. Write a function to print PASCAL Triangle. (TSRN)

```
#include <stdio.h>

int factorial(int num)
{
    long long fact = 1;
    for (int i = 1; i <= num; i++)
    {
        fact = fact * i;
    }
    return fact;
}


void print_pascals_triangle(int rows)
    {
        for (int i = 0; i < rows; i++)
            {
                for (int j = 0; j < rows - i - 1; j++)
                    {
                        printf(" ");
                    }
        for (int k = 0; k <= i; k++)
        {
            int binomial_coefficient = factorial(i) / (factorial(k) * factorial(i - k));
            printf("%d ", binomial_coefficient);
        }
        printf("\n");
    }
}

int main()
{
    int rows;
    printf("Enter the number of rows for Pascal's Triangle: ");
    scanf("%d", &rows);

    print_pascals_triangle(rows);

    return 0;
}


```

9. Write a function in C to merge two arrays of the same size sorted in descending order.

```

#include <stdio.h>

void mergeArrays(int arr1[], int n1, int arr2[], int n2, int merged[])
{
    for (int i = 0; i < n1; i++)
    {
        merged[i] = arr1[i];
    }
    for (int i = 0; i < n2; i++)
    {
        merged[n1 + i] = arr2[i];
    }

    int size = n1 + n2;
    int temp;

    // Sort the merged array using bubble sort
    for (int i = 0; i < size - 1; i++)
    {
        for (int j = 0; j < size - i - 1; j++)
        {
            if (merged[j] > merged[j + 1])
            {
                temp = merged[j];
                merged[j] = merged[j + 1];
                merged[j + 1] = temp;
            }
        }
    }
}

int main()
{
    int arr1[] = {1, 6, 3, 4};
    int arr2[] = {5, 5, 7, 9};
    int n1 = sizeof(arr1) / sizeof(arr1[0]);
    int n2 = sizeof(arr2) / sizeof(arr2[0]);

    int merged[n1 + n2];
    mergeArrays(arr1, n1, arr2, n2, merged);

    // Print the merged and sorted array
    for (int i = 0; i < n1 + n2; i++)
    {
        printf("%d ", merged[i]);
    }
    
    return 0;
}

```

```
#include<stdio.h>
void sort(int *ptr, int size)
{
    int i, j, temp;
    for (i = 0; i < size - 1; i++)
        {
            for (j = 0; j < size - i - 1; j++)
                {
                    if (ptr[j] > ptr[j + 1])
                        {
                            temp = ptr[j];
                            ptr[j] = ptr[j + 1];
                            ptr[j + 1] = temp;
                        }
                }
        }
}

int main()
{
    int size, i;
    printf("enter the random size for array: \n");
    scanf("%d",&size);
    int array[size];
    for(i=0; i<size; i++)
        {
            printf("enter array[%d]: \n",i);
            scanf("%d",&array[i]);
        }
    sort(array, size);
    printf("array in decendeing order:\n");
    
    for (i = 0; i < size; i++)
        {
            printf("%d ", array[i]);
        }
    
    return 0;
}
```
